```python
!pip install -q tensorflow matplotlib seaborn scikit-learn

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow import keras
from tensorflow.keras import layers, callbacks
from sklearn.metrics import confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')

print(f'TensorFlow: {keras.__version__}')
print('All imports OK')

# ===================================================================
# LOAD AND VISUALIZE DATA
# ===================================================================
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Show raw samples
fig, axes = plt.subplots(2, 10, figsize=(20,4))
fig.suptitle('MNIST Raw Data — 28x28 Pixel Handwritten Digits', fontsize=13, fontweight='bold')
for i, ax in enumerate(axes.flat):
    ax.imshow(x_train[i], cmap='gray')
    ax.set_title(f'Label: {y_train[i]}', fontsize=9)
    ax.axis('off')
plt.tight_layout()
plt.savefig('mnist_samples.png', dpi=150, bbox_inches='tight')
plt.show()

# Class distribution
fig, ax = plt.subplots(figsize=(10,4))
unique, counts = np.unique(y_train, return_counts=True)
bars = ax.bar(unique, counts, color=plt.cm.tab10.colors)
ax.set_title('Training Set Class Distribution', fontsize=12)
ax.set_xlabel('Digit')
ax.set_ylabel('Count')
for bar, c in zip(bars, counts):
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+50, str(c), ha='center', fontsize=9)
plt.tight_layout()
plt.savefig('class_distribution.png', dpi=150, bbox_inches='tight')
plt.show()

# ===================================================================
# PREPROCESS DATA
# ===================================================================
# Flatten and normalize
x_tr = x_train.reshape(-1, 784).astype('float32') / 255.0
x_te = x_test.reshape(-1, 784).astype('float32') / 255.0

# One-hot encode labels
y_tr = keras.utils.to_categorical(y_train, 10)
y_te = keras.utils.to_categorical(y_test, 10)

print(f'Train: {x_tr.shape}  |  Test: {x_te.shape}')
print('Pixel values normalised to [0.0 – 1.0]')

# ===================================================================
# BUILD MODEL
# ===================================================================
keras.backend.clear_session()

nn = keras.Sequential([
    layers.Input(shape=(784,), name='Input_784'),
    layers.Dense(512, name='Dense_512'),
    layers.BatchNormalization(name='BatchNorm_1'),
    layers.Activation('relu', name='ReLU_1'),
    layers.Dropout(0.3, name='Dropout_30pct'),
    layers.Dense(256, name='Dense_256'),
    layers.BatchNormalization(name='BatchNorm_2'),
    layers.Activation('relu', name='ReLU_2'),
    layers.Dropout(0.2, name='Dropout_20pct'),
    layers.Dense(128, activation='relu', name='Dense_128'),
    layers.Dense(10, activation='softmax', name='Output_Softmax'),
], name='MNIST_Classifier')

nn.compile(
    optimizer=keras.optimizers.Adam(0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

nn.summary()
print(f'Total parameters: {nn.count_params():,}')

# ===================================================================
# TRAIN MODEL
# ===================================================================
print('Training... (~3-5 minutes)')

history = nn.fit(
    x_tr, y_tr,
    batch_size=128,
    epochs=25,
    validation_split=0.1,
    callbacks=[
        callbacks.EarlyStopping(
            monitor='val_loss',
            patience=4,
            restore_best_weights=True,
            verbose=1
        ),
        callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=2,
            min_lr=1e-7,
            verbose=1
        ),
    ],
    verbose=1
)

print(f'Best val accuracy: {max(history.history["val_accuracy"])*100:.2f}%')

# ===================================================================
# PLOT TRAINING HISTORY
# ===================================================================
ep = range(1, len(history.history['accuracy'])+1)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))
fig.suptitle('Training History', fontsize=16, fontweight='bold')

# Accuracy plot
ax1.plot(ep, history.history['accuracy'], 'o-', color='#2196F3', lw=2.5, ms=5, label='Train')
ax1.plot(ep, history.history['val_accuracy'], 's--', color='#4CAF50', lw=2.5, ms=5, label='Val')
ax1.fill_between(ep, history.history['accuracy'], history.history['val_accuracy'], alpha=0.08, color='gray')
ax1.set_title('Accuracy per Epoch')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_ylim([0.85, 1.0])

# Loss plot
ax2.plot(ep, history.history['loss'], 'o-', color='#F44336', lw=2.5, ms=5, label='Train')
ax2.plot(ep, history.history['val_loss'], 's--', color='#FF9800', lw=2.5, ms=5, label='Val')
ax2.set_title('Loss per Epoch')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('training_curves.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved training_curves.png')

# ===================================================================
# EVALUATE MODEL
# ===================================================================
y_probs = nn.predict(x_te, verbose=0)
y_pred = np.argmax(y_probs, axis=1)
cm = confusion_matrix(y_test, y_pred)

# Confusion matrix
plt.figure(figsize=(13, 11))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=[f'Pred {i}' for i in range(10)],
    yticklabels=[f'True {i}' for i in range(10)],
    linewidths=0.4,
    linecolor='#ccc',
    cbar_kws={'label': 'Count'}
)
plt.title('Confusion Matrix — 10,000 Test Samples', fontsize=14, fontweight='bold', pad=15)
plt.ylabel('Actual Digit', fontsize=12)
plt.xlabel('Predicted Digit', fontsize=12)
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=150, bbox_inches='tight')
plt.show()

# Most confused pairs
print('Most confused digit pairs:')
cm2 = cm.copy()
np.fill_diagonal(cm2, 0)
for _ in range(5):
    i, j = np.unravel_index(cm2.argmax(), cm2.shape)
    print(f'  True {i} predicted as {j}: {cm2[i,j]} times')
    cm2[i, j] = 0

# Classification report
print('\n' + '='*70)
print('CLASSIFICATION REPORT')
print('='*70)
print(classification_report(
    y_test, y_pred,
    target_names=[f'Digit {i}' for i in range(10)],
    digits=4
))

# Overall metrics
loss, acc = nn.evaluate(x_te, y_te, verbose=0)
print('='*70)
print(f'FINAL TEST ACCURACY : {acc*100:.2f}%')
print(f'FINAL TEST LOSS     : {loss:.5f}')
print(f'CORRECT             : {int(acc*10000):,} / 10,000')
print(f'WRONG               : {10000-int(acc*10000):,} / 10,000')
print('='*70)

# Per-class accuracy bar chart
pca = cm.diagonal() / cm.sum(axis=1)
fig, ax = plt.subplots(figsize=(12, 5))
bars = ax.bar(range(10), pca*100, color=[plt.cm.RdYlGn(v) for v in pca])
ax.set_xticks(range(10))
ax.set_xticklabels([f'Digit {i}' for i in range(10)])
ax.set_ylim([96, 100])
ax.set_ylabel('Accuracy %')
ax.set_title('Per-Digit Test Accuracy', fontsize=13, fontweight='bold')
ax.axhline(acc*100, color='navy', linestyle='--', lw=1.5, label=f'Overall: {acc*100:.2f}%')
ax.legend()
for bar, v in zip(bars, pca):
    ax.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height() - 0.25,
        f'{v*100:.1f}%',
        ha='center',
        va='top',
        fontsize=9
    )
plt.tight_layout()
plt.savefig('per_class_accuracy.png', dpi=150, bbox_inches='tight')
plt.show()

# ===================================================================
# VISUALIZE PREDICTIONS
# ===================================================================
# 5x5 correct/wrong grid
fig, axes = plt.subplots(5, 5, figsize=(14, 14))
fig.suptitle(
    '25 Sample Predictions   Green=Correct   Red=Wrong',
    fontsize=14,
    fontweight='bold',
    y=1.01
)

for i, ax in enumerate(axes.flat):
    ok = y_pred[i] == y_test[i]
    conf = y_probs[i][y_pred[i]] * 100
    ax.imshow(x_test[i], cmap='gray_r')
    ax.set_title(
        f'Pred:{y_pred[i]} ({conf:.0f}%)\nTrue:{y_test[i]}',
        color='#2E7D32' if ok else '#C62828',
        fontsize=9,
        fontweight='bold'
    )
    for sp in ax.spines.values():
        sp.set_edgecolor('#2E7D32' if ok else '#C62828')
        sp.set_linewidth(3)
    ax.axis('off')

plt.tight_layout()
plt.savefig('predictions_grid.png', dpi=150, bbox_inches='tight')
plt.show()

# Misclassified examples
wrong = np.where(y_pred != y_test)[0]
print(f'\nTotal misclassified: {len(wrong)} / 10,000')

if len(wrong) >= 10:
    fig, axes = plt.subplots(2, 5, figsize=(14, 6))
    fig.suptitle(
        'Misclassified — Where the Model Was Wrong',
        fontsize=13,
        fontweight='bold'
    )

    for i, ax in enumerate(axes.flat):
        idx = wrong[i]
        conf = y_probs[idx][y_pred[idx]] * 100
        ax.imshow(x_test[idx], cmap='gray_r')
        ax.set_title(
            f'True:{y_test[idx]}  Pred:{y_pred[idx]}\n{conf:.0f}% confident',
            color='#C62828',
            fontsize=9,
            fontweight='bold'
        )
        ax.axis('off')

    plt.tight_layout()
    plt.savefig('misclassified.png', dpi=150, bbox_inches='tight')
    plt.show()
else:
    print('Not enough misclassifications to display (model is too accurate!)')

# ===================================================================
# SUMMARY
# ===================================================================
print('\n' + '='*70)
print('All done! Files saved:')
print('='*70)
for f in [
    'mnist_samples.png',
    'class_distribution.png',
    'training_curves.png',
    'confusion_matrix.png',
    'per_class_accuracy.png',
    'predictions_grid.png',
    'misclassified.png'
]:
    print(f'  ✓ {f}')
print('='*70)
```
